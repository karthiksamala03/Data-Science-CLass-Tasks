### 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 

```python
def healthy_person(a):
    if a > 2200:
        return 1
    else :
        return 0
```

```python
df['Health_Flag'] = df['Calories'].apply(healthy_person)
```

```python
df[df['Health_Flag'] == 1]['Id'].unique()
```

### 9. how many person are not a regular person with respect to activity try to find out those

```python
df.head()
```

```python
m = df[["ActivityDate","Id"]].groupby("Id").count().reset_index()
def lazy(a):
  for i in m["ActivityDate"]:
    if a <= 10:
      return"Inactive"
    elif 10 < a <= 20:
      return"Active"
    elif a > 20:
      return"Very Active"

m["Days"] = m["ActivityDate"].apply(lazy)
m
```

### 10. who is the third most active person in this dataset find out those in pandas and in sql both

```python
o = df[["TotalDistance","Id"]].groupby("Id").sum().reset_index()
p = o.sort_values(by = "TotalDistance", ascending=False,ignore_index=True)
p.iloc[:3]
```

### 11. who is the 5th most laziest person avilable in dataset find it out
